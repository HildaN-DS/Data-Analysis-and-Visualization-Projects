## Data Cleaning

- Format column names to be consistent
- Change data types

```python
# Rename the first column to 'Time' and convert to datetime
exchange_rates.rename(columns={exchange_rates.columns[0]: "Time"}, inplace=True)
exchange_rates["Time"] = pd.to_datetime(exchange_rates["Time"], errors="coerce")

# Rename other columns: replace spaces with underscores (excluding 'Time')
exchange_rates.columns = ["Time"] + [
    col.replace("[", "").replace("]", "").strip().replace(" ", "_")
    for col in exchange_rates.columns[1:]
]
# Convert all columns (except 'Time') to float
for col in exchange_rates.columns[1:]:
    exchange_rates[col] = pd.to_numeric(exchange_rates[col], errors="coerce")

# check result
print(exchange_rates.dtypes)
```

## Exploratory Data Visualizations

Generate some visualizations to help us explore, understand, and generate insights from the data.

We will use the `rolling mean` to see the trends of some currencies against the Euro

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

**Euro-US_dollar Trend**

```python
# Define major currency to analyze
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()

# Drop rows where US_dollar is NaN
euro_to_dollar.dropna(subset=['US_dollar'], inplace=True)

plt.figure(figsize=(9,6))

plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.title('Original values', weight='bold')

for i, rolling_mean in zip([2, 3, 4, 5, 6],
                           [7, 30, 50, 100, 365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'],
             euro_to_dollar['US_dollar'].rolling(rolling_mean).mean())
    plt.title('Rolling Window:' + str(rolling_mean), weight='bold')

plt.tight_layout() # Auto-adjusts the padding between subplots
plt.show()
```

### Inferences from Euro to US Dollar Exchange Rate

- From ~2002 to ~2008, the Euro generally appreciated against the Dollar.

- Post-2008, there's a decline and more volatility

**Euro-Sterling_pound Trend**

```python
euro_to_pound = exchange_rates[['Time', 'UK_pound_sterling']].copy()

# Drop rows where UK_pound_sterling is NaN
euro_to_pound.dropna(subset=['UK_pound_sterling'], inplace=True)

plt.figure(figsize=(9,6))

plt.subplot(3,2,1)
plt.plot(euro_to_pound['Time'], euro_to_pound['UK_pound_sterling'])
plt.title('Original values', weight='bold')

for i, rolling_mean in zip([2, 3, 4, 5, 6],
                           [7, 30, 50, 100, 365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_pound['Time'],
             euro_to_pound['UK_pound_sterling'].rolling(rolling_mean).mean())
    plt.title('Rolling Window:' + str(rolling_mean), weight='bold')

plt.tight_layout() # Auto-adjusts the padding between subplots
plt.show()
```

### Inferences from Euro to UK Pound Sterling Exchange Rate

- The Euro was weaker than the Pound in the early 2000s.
- There was a sharp appreciation in the Euro's value around 2008–2009, coinciding with the global financial crisis, where the Pound weakened.
- Post-2016 shows noticeable volatility, aligning with Brexit uncertainty and transition.

## Storytelling Data Visualization

To craft a compelling visual story from the Euro exchange rate data, I will anchor the narrative around key global and regional events that significantly impacted currency movements such as:

  - 2008 Financial Crisis

  - 2010–2012 Eurozone crisis

  - 2016–2020 Brexit and Trump era

  - 2020 COVID-19 pandemic

The goal is to connect financial data to real-world events, making the story relatable and engaging

### **2007 - 2009 Financial Crisis**

This crisis was a worldwide economic shock triggered by the collapse of the housing market in the United States, leading to the failure of major financial institutions and a severe credit crunch. Its effects rippled across global economies, particularly those with strong financial sectors or heavy exposure to international trade like the United Kingdom and Japan.

Exchange rates responded swiftly: as investors rushed to the U.S. dollar for safety, the Euro and other currencies saw sharp fluctuations.

```python
# Filter for the relevant time period
crisis_period = exchange_rates[(exchange_rates['Time'] >= '2006-01-01') &
                               (exchange_rates['Time'] <= '2011-12-31')].copy()
crisis_period.set_index('Time', inplace=True)

# Select the currencies of interest
currencies = {
    'US_dollar': 'US Dollar',
    'UK_pound_sterling': 'UK Pound',
    'Indian_rupee': 'Indian Rupee',
    'Japanese_yen': 'Japanese Yen',
    'Swiss_franc': 'Swiss Franc',
    'Chinese_yuan_renminbi': 'Chinese Yuan'
}

# Set up subplots
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(12, 8))
axs = axs.flatten()

# Plot each currency
for i, (col, label) in enumerate(currencies.items()):
    data = crisis_period[[col]].copy().dropna()
    data['RollingMean'] = data[col].rolling(window=30).mean()

    axs[i].plot(data.index, data['RollingMean'], color='blue', label='30-day Rolling Mean')
    axs[i].set_title(f'Euro to {label} (2006–2011)', fontsize=10, weight='bold')

    # Highlight crisis window: from Jan 2008 to Dec 2009
    axs[i].axvspan(pd.Timestamp('2008-01-01'), pd.Timestamp('2009-12-31'),
                   color='red', alpha=0.2, label='Crisis Period')

    axs[i].grid(True)
    axs[i].legend(fontsize=8)

# Layout
plt.suptitle('Impact of the 2007–2008 Financial Crisis on Euro Exchange Rates (30-day Rolling Mean)',
             fontsize=12, weight='bold')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

### Inferences

📉 Euro to US Dollar:
The Euro peaked in mid-2008, then dropped sharply as investors flocked to the dollar for safety during the crisis.

📉 Euro to UK Pound:
The Euro surged against the pound, reflecting the UK’s deep banking exposure and economic vulnerability.

📈 Euro to Indian Rupee:
The Euro appreciated as capital fled emerging markets; the Rupee showed post-crisis volatility and a partial rebound.

📉 Euro to Japanese Yen:
The Euro fell steeply, highlighting the Yen’s role as a safe-haven currency during financial stress.

📉 Euro to Swiss Franc:
The Euro depreciated as investors turned to the stable Swiss Franc, which continued strengthening post-crisis.

📉 Euro to Chinese Yuan:
The Euro declined while the Yuan stayed stable, reflecting China’s exchange controls and economic resilience.

### **2016–2020 Brexit and Trump era**

Time Window: 2015–2021 (to capture lead-up and after-effects)

Key Events:

- June 2016: Brexit referendum

- January 2017: Trump inauguration

- 2018–2019: US-China trade tensions

```python
# Filter for this period
brexit_trump_period = exchange_rates[(exchange_rates['Time'] >= '2015-01-01') &
                                     (exchange_rates['Time'] <= '2020-01-01')].copy()
brexit_trump_period.set_index('Time', inplace=True)

# Define currencies of interest
currencies = {
    'US_dollar': 'US Dollar',
    'UK_pound_sterling': 'UK Pound',
    'Chinese_yuan_renminbi': 'Chinese Yuan',
    'Japanese_yen': 'Japanese Yen',
    'Indian_rupee': 'Indian Rupee',
    'Swiss_franc': 'Swiss Franc'
}

# Set up subplots
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(14, 8))
axs = axs.flatten()

# Plot each currency
for i, (col, label) in enumerate(currencies.items()):
    if col not in brexit_trump_period.columns:
        continue
    data = brexit_trump_period[[col]].copy().dropna()
    data['RollingMean'] = data[col].rolling(window=30).mean()

    axs[i].plot(data.index, data['RollingMean'], color='blue', label='30-day Rolling Mean')
    axs[i].set_title(f'Euro to {label} (2015–2021)', fontsize=10, weight='bold')

    # Highlight major events
    axs[i].axvline(pd.Timestamp('2016-06-23'), color='red', linestyle='--', linewidth=1, label='Brexit Vote')
    axs[i].axvline(pd.Timestamp('2017-01-20'), color='orange', linestyle='--', linewidth=1, label='Trump Inauguration')
    axs[i].axvspan(pd.Timestamp('2018-07-01'), pd.Timestamp('2019-12-31'), color='green', alpha=0.2, label='US–China Trade Tensions')


    axs[i].grid(True)
    axs[i].legend(fontsize=8)

# Layout
plt.suptitle('Impact of Brexit and Trump Era on Euro Exchange Rates (30-day Rolling Mean)',
             fontsize=14, weight='bold')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

**🇬🇧 Brexit Vote (June 2016)**

Euro to UK Pound: Sharp appreciation of the Euro after the Brexit vote, reflecting market shock and weakened confidence in the British economy.

Euro to US Dollar / Yen / Swiss Franc: Brief dips or volatility, indicating broader uncertainty but less severe than with GBP.

Euro to Indian Rupee / Chinese Yuan: Mild fluctuations, minimal immediate impact, suggesting limited exposure.

**🇺🇸 Trump Inauguration (January 2017)**

Euro to US Dollar: Initially steady, then a rise through 2017–2018 as markets reacted to fiscal policy expectations and tax reforms.

Euro to Chinese Yuan / Indian Rupee: Gradual Euro gains, hinting at growing global uncertainty and capital shifts toward the Euro.

Euro to Swiss Franc / Yen: Mixed trends; temporary Euro strength in 2017, followed by reversals in 2018.

**⚔️ US–China Trade Tensions (2018–2019)**

Euro to US Dollar: Decline in Euro post-2018, reflecting investor movement toward the dollar as a safer asset.

Euro to Chinese Yuan: Euro gained ground, while the Yuan remained within a controlled range, showing Beijing's policy-driven stability.

Euro to Japanese Yen / Swiss Franc: Euro weakened significantly, highlighting a shift to traditional safe-haven currencies.

Euro to Indian Rupee: Euro appreciated sharply, reflecting pressure on emerging markets amid global trade uncertainty.

### **2020 COVID-19 pandemic**

```python
import matplotlib.pyplot as plt
import pandas as pd

# Ensure 'Time' column is datetime
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])

# Filter for the COVID-19 and post-COVID period
covid_period = exchange_rates[(exchange_rates['Time'] >= '2019-01-01') &
                              (exchange_rates['Time'] <= '2022-12-31')].copy()
covid_period.set_index('Time', inplace=True)

# Define currencies of interest
currencies = {
    'US_dollar': 'US Dollar',
    'UK_pound_sterling': 'UK Pound',
    'Indian_rupee': 'Indian Rupee',
    'Japanese_yen': 'Japanese Yen',
    'Swiss_franc': 'Swiss Franc',
    'Chinese_yuan_renminbi': 'Chinese Yuan'
}

# Set up subplots
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(14, 8), sharex=True)
axs = axs.flatten()

# Plot each currency
for i, (col, label) in enumerate(currencies.items()):
    if col in covid_period.columns:
        data = covid_period[[col]].copy().dropna()
        data['RollingMean'] = data[col].rolling(window=30).mean()

        axs[i].plot(data.index, data['RollingMean'], color='blue', label='30-day Rolling Mean')
        axs[i].set_title(f'Euro to {label} (2019–2022)', fontsize=10, weight='bold')

        # Highlight COVID period: March 2020 to December 2021
        axs[i].axvspan(pd.Timestamp('2020-03-01'), pd.Timestamp('2021-12-31'),
                       color='red', alpha=0.3, label='COVID-19 Period')

        axs[i].grid(True)
        axs[i].legend(fontsize=8)
        axs[i].tick_params(axis='x', rotation=45)

# Layout
plt.suptitle('Impact of COVID-19 on Euro Exchange Rates (30-day Rolling Mean)',
             fontsize=14, weight='bold')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

#### Inferences

📉 Euro to US Dollar: The Euro rose early in the pandemic but fell sharply post-2021 as the dollar strengthened with U.S. rate hikes.

📉 Euro to UK Pound: Initial gains gave way to a decline, reflecting the UK’s recovery momentum and Eurozone struggles.

📈 Euro to Indian Rupee: The Euro surged mid-pandemic as the Rupee weakened, then corrected with India’s economic rebound.

📈 Euro to Japanese Yen: The Euro steadily rose, highlighting Japan’s ultra-loose monetary policy and Yen weakness.

📉 Euro to Swiss Franc: The Euro weakened consistently, reflecting the Franc’s safe-haven appeal and Swiss stability.

📉 Euro to Chinese Yuan: The Euro declined from 2021 as China’s strong recovery and policy controls supported the Yuan.

## 📘 Conclusion:
The Euro’s journey through the financial crisis, Brexit, and COVID-19 reveals its role as a barometer of global sentiment— **rising amid localised shocks but retreating during systemic crises.**

Whether reacting to capital flight during financial turmoil, political upheaval from Brexit and Trump-era policies, or shifting monetary responses during the pandemic, the Euro has mirrored global trust and uncertainty. **It tends to strengthen against vulnerable or emerging markets but weakens against safe-haven currencies like the Dollar, Yen, and Swiss Franc during global stress.**
